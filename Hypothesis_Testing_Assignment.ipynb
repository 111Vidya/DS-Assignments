Q1.A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.
minitabfile : Cutlets.csv

Null Hypothesis Ho - Diameter of the cutlets from both units are SAME.
Alternate Hypothesis H1 - Diameter of the cutlets from both units are NOT SAME

```python
import pandas as pd
ct=pd.read_csv("Cutlets.csv")
ct.head()
```

```python
ct.shape
```

```python
ct.isnull()
```

```python
#Two Sample T Test
import scipy
from scipy import stats
```

```python
stats.ttest_ind(ct['Unit A'],ct['Unit B'])
```

**ANS : PValue = 0.47 which is greater than 0.05.
PValue > 0.05 accept Null Hypothesis Ho.
Diameter of the cutlets from both units are SAME.**

Q2. A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.       
Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

```python
lb=pd.read_csv("LabTAT.csv")
lb
```

```python
lb.isnull()
```

```python
lb.isnull().sum()
```

```python
lb.mean()
```

Null Hypothesis Ho - TAT of all Labs are same TATL1=TATL2=TATL3=TATL4.
Alternate Hypothesis H1 - TAT of atleast one of the lab NOT SAME

```python
stats.f_oneway(lb.iloc[:,0],lb.iloc[:,1],lb.iloc[:,2],lb.iloc[:,3])
```

**ANS: PValue < 0.05 accept Alternate Hypothesis H1.
TAT of atleast one Laboratory is different**

Q3. Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

Null Hypothesis Ho -All Proportions are equal.
Alternate Hypothesis H1 -Not all proportions are equal


```python
br=pd.read_csv("BuyerRatio.csv")
br
```

```python
br1=br.drop(columns="Observed Values",axis=1)
br1
```

```python
Chisquares_results=scipy.stats.chi2_contingency(br1)
Chisquares_results
```

**ANS: PValue > 0.05 accept Null Hypothesis Ho
All Proportions are equal**
