```python
!pip install mlxtend
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np
```

```python
bk=pd.read_csv("book.csv")
bk
```

```python
#Apriori Algorithm
frequent_itemsets = apriori(bk, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
#calc Lift Ratio and provide data set with lift ratio>1
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

We can identify 5 association rules from the below model
*   Obs 28 if ItalCook is bought, then CookBks is bought with a cofidence of 100%
*   Obs 77 if ChildBks,ArtBks is bought, GeogBks is bought with a confidence of 63%
*   Obs 84 if DoitYBks,CookBks is bought, ArtBks is bought with confidence of 54%
*   Obs 95 if ArtBks,CookBks is bought, GeogBks is bought with confidence of 62%
*   Obs 96 if GeogBks,CookBks is bought, ArtBks is bought with confidence of 54%








```python
#Sort Data in decending order and provide top 10
rules.sort_values('lift',ascending = False).head(10)
```
