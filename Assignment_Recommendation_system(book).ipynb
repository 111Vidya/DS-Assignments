Build a recommender system by using cosine similarties score.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bk=pd.read_csv("/content/bookrs.csv",encoding="unicode_escape")
bk
```

```python
bk1=bk.drop(["Unnamed: 0"],axis=1)
bk1
```

```python
bk1.shape
```

```python
bk1.info()
```

```python
bk1.isnull().sum()
```

```python
bk1["User.ID"].duplicated().sum()
```

```python
bk1=bk1.sort_values(by="User.ID").reset_index(drop=True)
bk1
```

```python
#Unique User ID
len(bk1["User.ID"].unique())
```

```python
#Unique books
len(bk1["Book.Title"].unique())
```

```python
#User based collaborative filterngs.User item interaction matrix.
bk2=bk1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
bk2
```

```python
#replace index with user ID unque values
bk2.index=bk1['User.ID'].unique()
bk2
```

```python
#Fill NA values with 0
bk2.fillna(0,inplace=True)
bk2
```

```python
from sklearn.metrics import pairwise_distances
```

```python
bk3 = 1 - pairwise_distances( bk2.values,metric='cosine')
bk3
```

```python
bk3_df = pd.DataFrame(bk3)
```

```python
#Set the index and column names to user ids
bk3_df.index = bk["User.ID"].unique()
bk3_df.columns = bk["User.ID"].unique()
bk3_df
```

```python
#Diagonal information to be replaced with 0
np.fill_diagonal(bk3, 0)
bk3_df
```

```python
#Top 10 similar IDs with high Cosine value
bk3_df.idxmax(axis=1)[0:10]
```
